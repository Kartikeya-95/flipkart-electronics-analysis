# Analysis of products in different categories with some visualizations. <br>
<b>Note :</b> Since the exact number of sales data is not available, so we assume that number of ratings represent minimum sales of different products. It comes from the assumption that people who rated the product must have also bought it.

## 1. Checking category wise top 5 selling products.

```python
cat = df['category'].unique()
cat
```

```python
# Air Conditioner Segment
top_5 = df[df['category'] == cat[0]].sort_values(by = 'num_ratings', ascending = False).head()

sns.barplot(x = 'num_ratings', y = 'description', data = top_5, orient = 'h')
plt.xlabel('Minimum number of sales', fontdict = {'fontsize' : 12}, labelpad = 10)
plt.ylabel('Product Description', fontdict = {'fontsize' : 12}, labelpad = 10)
plt.yticks(range(5), ['\n'.join(wrap(t, 60)) for t in top_5['description']])
plt.title('Top 5 selling products in {} segment'.format(cat[0]), fontdict = {'fontsize' : 16}, pad = 10)
plt.grid(b = '--', axis = 'x')

ypos = 0
for i in range(len(top_5['price'])):
    plt.text(x = 17000, y = ypos, s = u'\u20B9 ' + str(top_5['price'].iloc[i]))
    ypos = ypos + 1
    
plt.text(x = 21000, y = 2.3, s = 'Price', fontdict = {'fontsize' : 12}, rotation = 90)
plt.show()
print('Amount gathered by top 5 selling products (in crores) : \u20B9 {}'.format((top_5['num_ratings'] * top_5['price']).sum()/10000000))
```

```python
# There are duplicates in description column which means there might be different sellers of the same product or same product
# is captured more than once. We will drop duplicate rows based on the below mentioned subset. If any value in at least one 
# column from below subset is different then that row will be kept.

print('Total rows before removal of duplicates : {}'.format(len(df)))
df.drop_duplicates(subset = 
                   ['description', 'is_fassured', 'rating', 'discount (%)', 'price', 'mrp', 'num_ratings', 'num_reviews']
                   , inplace = True)
print('Total rows after removal of duplicates : {}'.format(len(df)))
```

## 2. Category-wise revenue analysis.

```python
df['Sales (in crores)'] = (df['num_ratings'] * df['price']) / 10000000
```

```python
total_sales = df[['category', 'Sales (in crores)']].groupby('category').sum()['Sales (in crores)'].sort_values(ascending = False)
```

```python
total_sales.plot(kind = 'barh', figsize = (10,6))
plt.xlabel('Total Sales \u20B9 (in crores)', fontdict = {'fontsize':12})
plt.ylabel('Category', fontdict = {'fontsize':12})
plt.title('Category-wise minimum total sales', fontdict = {'fontsize' : 16}, pad = 20)
plt.grid(b = '-', axis = 'x')
plt.show()
```

From above graph we can see that <b>Mobiles</b> are the primary source of revenue in Electronics category. Major part of revenue is generated by Mobile segment. <br> <br>
Note : <b>Number of ratings</b> is used as minimum threshold for number of sales. These figures will not depict actual sales. It just gives an idea of sales trend calculated using <b>number of ratings</b>. If there is no rating for a product, then it does not mean that product is not sold at all. 

## 3. Category-wise rating analysis.

```python
# Calculating category wise mean ratings excluding 0, as rating scale varies from 1 to 5. 0 value means no rating.

mean_rating = df[df['rating'] != 0][['category', 'rating']].groupby('category').mean()['rating'].sort_values(ascending = False)
```

```python
mean_rating.plot(kind = 'barh', figsize = (10,6))
plt.xlabel('Average Rating', fontdict = {'fontsize':12})
plt.ylabel('Category', fontdict = {'fontsize':12})
plt.title('Category-wise average ratings', fontdict = {'fontsize' : 16}, pad = 20)
plt.grid(b = '-', axis = 'x')
plt.show()
```

```python
# Grouping on the basis of category and brand of filtered data where rating != 0 and num_ratings > 5000.
best_brands = df[(df['rating'] != 0) & (df['num_ratings'] > 5000)][
    ['category', 'brand', 'description', 'rating']
].groupby(['category', 'brand']).mean().reset_index()

# Further filtering on the basis of max rating
top_brands = {
    'category' : [],
    'brand' : [],
    'max_avg_rating' : []
}
for i in cat:
    x = best_brands[(best_brands['category'] == i)]
    x = x[x['rating'] == x['rating'].max()]
    if len(x['brand']) != 0:
        top_brands['category'].append(i)
        top_brands['brand'].append(x['brand'].values)
        top_brands['max_avg_rating'].append(x['rating'].iloc[0])
    
best_brands = pd.DataFrame(top_brands)
```

```python
plt.figure(figsize = (12,8))
sns.barplot(x = 'max_avg_rating', y = 'category', data = best_brands, orient = 'h', color = '#327bb3')
for i in np.arange(4,5,0.2):
    plt.axvline(x = i, ls = '--', lw = 0.5, color = 'black')
    
for i in range(len(best_brands['brand'])):
    plt.text(x = 1, y = 0.2 + i, s = 'Brand : ' + str(list(best_brands['brand'].iloc[i])), color = 'white')
    
plt.title('Category-wise best performing brands based on rating', fontdict = {'fontsize' : 16}, pad = 20)
plt.xlabel('Maximum Average Rating', fontdict = {'fontsize' : 12}, labelpad = 10)
plt.ylabel('Category', fontdict = {'fontsize' : 12}, labelpad = 10)

plt.savefig('../Category_wise_best_performing_brands.png', bbox_inches = 'tight')
plt.show()
```

The above graph shows best performing brands in their respective category based on maximum average rating. Here, products with number of ratings > 5000 have been considered to aggregate the rating value. This filtering is done because there are products with average rating of 5. This is because there are some products having very less number of ratings and all of them are 5. Thus, a threshold of 5000 number of ratings have been set to get an unbiased aggregate value of rating. Also, the brand names which come out from this analysis makes sense as they are all very popular brands. <br> <br>
Note : <b> Desktop PCs </b> category is not there, as no product in that category has greater than 5000 number of ratings. In fact, the maximum number of ratings for a product in this segment is <b> 699 </b>.

## 6. Category-wise best performing brands, based on total sales.
